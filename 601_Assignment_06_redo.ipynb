```python
import openpyxl as xl
from openpyxl.workbook import workbook as wb
from openpyxl.utils.dataframe import dataframe_to_rows
import os
import pandas as pd
```

```python
# load excel file

path_to_file = os.path.join(os.getcwd(), 'week_05_homework_XLSX_openpyxl.xlsx')
wb = xl.load_workbook(path_to_file, data_only= True)
```

```python
# assign worksheets 

main = wb['main']
another = wb['another']
```

```python
# create output file

output_file = os.path.join(os.getcwd(), 'output_file.xlsx')
wb2 = xl.Workbook()
ws = wb2.active
ws.title = "Output"
output = wb2["Output"]
wb2.save(output_file)
```

```python
# function to extract column names

def col_names(x):
    columns = []
    for col in x.iter_cols(max_row= 1):
        for cell in col:
            headers = cell.value
            columns.append(headers)
    return columns
```

```python
# function to create dataframes of each worksheets

def create_df(y):
    df = pd.DataFrame(y.values)
    df.columns = col_names(y)
    df = df[1:]
    return df
```

```python
# create dataframes of each worksheets

df1 = create_df(main)
df2 = create_df(another)
```

```python
# reset the column name for "another" worksheet

df2.rename(columns = {"p_id":"patient id"}, inplace = True)
```

```python
# merge the dataframes

merged_df = pd.merge(df1, df2, how = "outer", on ="patient id")
merged_df
```

```python
# export the results to new Excel file

for r in dataframe_to_rows(merged_df, index = False, header = True):
    output.append(r)
wb2.save(output_file)
```
